```python
df_padded = df.fillna(0)
t = df_padded.iloc[:, 0:9000]
y = df_padded.pop('y')
print(t.shape)
print(y.shape)
```

```python
def segment_signal(target_length, dataframe):
    notna = dataframe.notna().sum(axis=1)
    orig_lengths = np.zeros(len(notna)) + notna
    col_names = list(dataframe.columns)
    col_names = col_names[0:target_length]
    collect_samples = list()
    for row in range(dataframe.shape[0]):
        sample = dataframe.iloc[row, :] # get row as a Series
        sample = sample.dropna()
        sample_length = sample.shape[0] # we're handling a Series not a Dataframe
        while sample_length < target_length:
            duplicated_sample = sample.append(sample)
            sample = duplicated_sample
            sample_length = sample.shape[0] # we're handling a Series not a Dataframe
        training_sample = sample[0:target_length]
        training_sample.index = col_names # index bc. handling a Series
        collect_samples.append(training_sample)
    finished_df = pd.DataFrame(data=collect_samples)
    return finished_df
```

```python
import warnings
warnings.filterwarnings('ignore')

rn = np.random.randint(1, 100, 1)

seq = t_scaled.iloc[rn]
seq = np.squeeze(seq)

x = np.array(range(0, 9000))

n = len(seq)
COMPONENTS = [10, 100, 500, n]

for c in COMPONENTS:
    plt.figure(figsize=(16,3))
    colors = np.linspace(start=100, stop=255, num=c)
    for i in range(c):
        Y = np.fft.fft(seq)
        np.put(Y, range(i+1, n), 0.0)
        ifft = np.fft.ifft(Y)
        plt.plot(x, ifft, color=plt.cm.Reds(int(colors[i])), alpha=.70)
    
    plt.title("First {c} fourier components".format(c=c))
    plt.plot(x,seq, label="Original dataset")
    plt.grid(linestyle='dashed')
    plt.legend()
    plt.show()
```

## Transformation to Spectrogram with STFT

```python
from scipy.signal import stft
import librosa
import imageio
import os, glob
```

```python
def spec(x):   # samplerate
    x = x.astype(float)
    amp = librosa.amplitude_to_db(np.abs(librosa.stft(x, n_fft=255, hop_length=1)), ref=np.max)
    return amp
```

At this point we want to save our spectrograms as .png images for feeding into the CNN later on.

```python
# create spec data in folder structure
def create_specs(data, labels, path):

    if not os.path.isdir(path): 
        os.makedirs(path)
    for cl in c_list:
        if not os.path.isdir(path + '/' + cl): 
            os.makedirs(path + '/' + cl)

    c1,c2,c3,c4 = 1,1,1,1

    for i, j in zip(range(len(data.index)), range(len(labels.index))):
        x = data.iloc[[i]]
        x = x.to_numpy()
        x = x[0,]
        amp = spec(x)
        if labels.values[j] == 0:
            c1 = save_img(amp, labels.values[j], c1)
        if labels.values[j] == 1:
            c2 = save_img(amp, labels.values[j], c2)
        if labels.values[j] == 2:
            c3 = save_img(amp, labels.values[j], c3)
        elif labels.values[j] == 3:
            c4 = save_img(amp, labels.values[j], c4)
```

```python
# save images
def save_img(amp, cl, cnt):
    plt.figure(figsize=(6, 3))
    plt.axis('off')
    plt.imshow(amp, aspect='auto')
    plt.savefig(os.path.join(path, c_list[cl], c_list[cl]) + '_%d.png' %cnt, bbox_inches='tight', pad_inches=0)
    plt.clf()
    plt.close()
    cnt += 1
    return cnt
```

# Train a convolutional neural network (CNN)

```python
# import libraries
from matplotlib.image import imread
import os, glob
import numpy as np
import random
from keras.utils import to_categorical
from sklearn.utils import shuffle
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import f1_score
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Conv2D, Flatten, Activation, MaxPooling2D, Dropout
from tensorflow.keras import optimizers
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.callbacks import ReduceLROnPlateau
```

```python
# define folder path
source_path = '/Users/victor/Desktop/Advanced_ML_ETH/task3/'
folder = 'spec_data'
path = os.path.join(source_path, folder)
print(path)
```

Extract classes

```python
c_list = os.listdir(path) 
c_list.sort()
print(c_list)
```

Import all images

```python
candidates = []
for filename in glob.iglob(path + '/**/*.png', recursive=True):
    candidates.append(filename)
random.shuffle(candidates) # shuffle images
print(len(candidates))
```

Let's choose on how much data we want to train our model.

```python
nb = 500 #  take 'len(candidates)' for all data
```

Extract image dimensions

```python
dim_img = imread(candidates[0])
dim_img = dim_img[:,:,:3]
img_height, img_width, img_depth = dim_img.shape
print(img_height, img_width, img_depth)
```

Extract weights

```python
cls0 = 3030
cls1 = 443
cls2 = 1474
cls3 = 170
total = cls0+cls1+cls2+cls3
w_c1 = (1 / cls0)*(total)/4.0 
w_c2 = (1 / cls1)*(total)/4.0
w_c3 = (1 / cls2)*(total)/4.0
w_c4 = (1 / cls3)*(total)/4.0
weights = {0:w_c1, 1:w_c2, 2:w_c3, 3:w_c4}
print(weights)
```

Prepare batchholders for images and labels

```python
X = np.zeros((len(candidates[0:nb]),img_height,img_width,img_depth))
y = np.zeros((len(candidates[0:nb])))
```

Fill batchholders according to classes and create training set

```python
for i in range(0,nb):
    split = os.path.basename(candidates[i]).split('_')
    class_type = split[0]
    image = imread(candidates[i])
    image = image[:,:,:3]
    X[i, :] = image
    y[i] = c_list.index(class_type)
print(X.shape, y.shape)
```

Define network settings

```python
input_shape = (img_height, img_width, img_depth)
nclasses = 4
epochs = 3
batch_size = 15 
loss_type = 'categorical_crossentropy' 
class_type = 'categorical' 
activation_fun = 'softmax' 
learning_rate = 0.00015
dropout_rate = 0.5
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2,
                              patience=4, min_lr=0.000005)
early = EarlyStopping(monitor='val_loss', patience=4, verbose=1, mode='min')

# merge callbacks
callbacks = [early, reduce_lr]
```

Design the CNN architecture

```python
# conv layer 1
model = Sequential()
model.add(Conv2D(16, (3, 3), input_shape=input_shape))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# conv layer 2
model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# conv layer 3
model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# conv layer 4
model.add(Conv2D(128, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# conv layer 5
model.add(Conv2D(256, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))

# flatten layer
model.add(Flatten())

# dense layer 1
model.add(Dense(800))
model.add(Activation('relu'))

# dense layer 2
model.add(Dense(800))
model.add(Activation('relu'))

# dense layer 3
model.add(Dense(800))
model.add(Activation('relu'))

# dense layer 4
model.add(Dense(800))
model.add(Activation('relu'))

# dropout layer
model.add(Dropout(dropout_rate))

# output layer
model.add(Dense(nclasses))
model.add(Activation(activation_fun))

opt = optimizers.Adam(lr=learning_rate, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
model.compile(loss=loss_type,
              optimizer=opt,
              metrics=['accuracy', tf.keras.metrics.Precision() , tf.keras.metrics.Recall()])

# print model architecture
print(model.summary())
```

We have to set a random seed for reproducability

```python
seed = 7
np.random.seed(seed)
```

Define 5-fold cross validation

```python
kfold = StratifiedKFold(n_splits=5, shuffle=True, random_state=seed)
```

Now let's train our model!

```python
fold = 1
f1 = []
for train, val in kfold.split(X, y):

    # one-hot encoding
    y_t = to_categorical(y[train])
    y_v = to_categorical(y[val])
    
    # train model
    history = model.fit(X[train], y_t, 
        epochs=epochs, 
        callbacks=callbacks, 
        batch_size=batch_size, 
        class_weight = weights,
        validation_data=(X[val],y_v))
    
    # calculate F1 score
    y_pred = model.predict_classes(X[val])
    F1 = f1_score(y[val], y_pred, average='micro')
    f1.append(F1)
    
    print('At fold:', fold, ' / F1 score:', F1)
    
    #model.save('T3_net' + str(fold) + '.h5')
    
    fold += 1


print('Mean F1 score: ', np.mean(f1), 'Std: ', np.std(f1))
```
